```python
from nltk.tokenize import sent_tokenize, word_tokenize
w_t=[]
f = codecs.open('my_corpus1.txt', encoding="utf-8")
lines = f.readlines()
for line in lines:
     z=word_tokenize(line)
     for w in z:
         w_t.append(w.lower())
def word_freq(word):
        c=0
        for w in w_t:
            if(w==word):
                c=c+1
        return c
```

```python
def ngrams(input, n):
    output = []
    for i in range(len(input)-n+1):
        output.append(input[i:i+n])
    return output
```

```python
def prob(u1,u,u2):
    p=[]
    for i in u1:
      c=0
      for j in u:
        if(i==j):
            c+=1
      m=c/len(u2)
      p.append(m)
    return p
```

```python
#add one smoothing
def prob_add_one(u1,u,u2,v):
    p=[]
    for i in u1:
      c=0
      for j in u:
        if(i==j):
            c+=1
      if(c==0):
        c=c+1
        v=v+1
      m=c/(len(u2)+v)
      p.append(m)
    return (p,v)
```

```python
v=0
pu,v=prob_add_one(u1,u,u,v)
pb,v=prob_add_one(b1,b,u,v)
pt,v=prob_add_one(t1,t,b,v)
print(pu)
print(pb)
print(pt)
```
